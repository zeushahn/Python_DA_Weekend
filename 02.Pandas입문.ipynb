#### List를 이용한 DataFrame 생성

```python
import pandas as pd
```

```python
# List로 Data 정의
data = [
	[15, '남', '덕영중'],
	[17, '여', '수리중']
]
data
```

```python
df = \
	pd.DataFrame(
		data,
		index=['준서',"예은"],
		columns=['나이','성별','학교']
	)
df
```

```python
# 나이 항목만 추출 : Index를 정의하여 이름이 보인다. 
df.loc[:,'나이']
```

```python
# 준서의 나이만 추출하기
print(df.loc['준서','나이'])
```

```python
# 예은의 나이, 학교를 추출
df.loc['예은',['나이','학교']]
```

---
#### 행과 열이름 변경하기

```python
# 행과 열이름 확인
print(df.index)
print(df.columns)
```

```python
# 행이 이름을 학생1, 학생2로 변경하기
df.index = ['학생1', '학생2']
df
```

```python
# 컬럼이름 : 연령, 남녀, 소속 으로 변경하기
df.columns = ['연령', '남녀', '소속']
df
```

#### 원하는 부분만 변경하기

```python
# 연령 -> 나이 
df.rename(
	columns={
		'연령':'나이'
	},
	inplace=True
)

df
```

```python
# rename 1번으로 남녀 -> 성별, 소속 -> 학교로 변경하기
df.rename(
	columns={
		'남녀':'성별',
		'소속':'학교'
	},
	inplace=True
)

df
```

```python
# 행이름 변경하기 :학생1 -> 준서, 학생2:예은 
df.rename(
	index={
		'학생1':'준서',
		'학생2':'예은'
	},
	inplace=True
)
df
```

```python
# Ex : 나이 -> 연령, 예은 -> 학생2 

df.rename(
	columns={
		'나이':'연령'
	},
	index={
		'예은':'학생2'
	},
	inplace=True
)

df
```

#### 행삭제

```python
exam_data = {
	'수학' : [90, 80, 70],
	'영어' : [98, 89, 95],
	'음악' : [85, 95, 100],
	'체육' : [100, 90, 90],
}
exam_data
```

```python
# DataFrame 생성하기
df = \
	pd.DataFrame(
		exam_data
	)
df
```

#### 데이터 추출하기 위한 선택
- loc : index나 columns의 이름을 기준으로 선택 
- iloc : 정수형 index를 기준으로 선택

```python
df
```

```python
# index가 서준인 행을 추출
df.loc['서준',:]
df.iloc[0,:]
```

```python
# 서준과 우현의 정보를 추출
df.loc[['서준','우현'],:] # indexing
df.loc['서준':'우현',:] # Slicing

df.iloc[[0,1],:] # indexing
df.iloc[0:1+1,:] # Slicing
```

#### Slicing의 step 사용
iloc[start:stop+1:step, start:stop+1:step]

```python
# 행은 모두 추출하고 열은 0, 2번으로 추출(단, Slicing만 사용가능)

df.iloc[:,::2] 
```
