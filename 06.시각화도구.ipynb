```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("../Data/시도별 전출입 인구수.xlsx")
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df.info()
```

```python
# 행번호 0을 삭제 
df.drop(
    0,
    axis='rows',
    inplace=True
)
df.head()
```

```python
# Index 재정리
df.reset_index(
    inplace=True,
    drop=True
)
df.head()
```

```python
# NaN 정리
# 현재 NaN을 상위의 값으로 채움

df.ffill(inplace=True) # bfill: 하위값으로 채우기 
df.head()
```

```python
df.info()
```

#### 년도별 컬럼에 대해 Object를 숫자로 변환

```python
# - -> 0으로 변경
df.replace(
    '-',
    0,
    inplace=True
)
df.head()
```

```python
df.info()
```

> 년도별 데이터가 시각화 할 수 있는 숫자로 변경 되었다.

---
#### 전출입 시각화

```python
# 서울에서 타지역으로 이동한 데이터만 추출
df_seoul = \
    df[(df.전출지별 == '서울특별시') & (df.전입지별 != '서울특별시')]

df_seoul.head()
```

```python
# 전출지별 컬럼 삭제
df_seoul.drop(
    '전출지별',
    axis='columns',
    inplace=True
)

df_seoul.head()
```

```python
# 전입지별 -> 전입지 변경
df_seoul.rename(
    {
        '전입지별' : '전입지'
    },
    axis='columns',
    inplace=True
)

df_seoul.head()
```

```python
# 전입지 컬럼을 Index로 변경
df_seoul.set_index(
    '전입지',
    inplace=True
)
df_seoul.head()
```

---
#### 서울에서 경기도로 이동한 인구의 선그래프로 시각화

```python
sr_one = df_seoul.loc['경기도',:]
sr_one.head()
```

```python
# 선그래프
plt.plot(
    sr_one
)
plt.show()
```

#### 부산광역시에서 서울특별시로 전입한 그래프

```python
df_busan = \
    df[(df.전출지별 == '부산광역시') & (df.전입지별 == '서울특별시')]

df_busan
```

```python
# 전출지별 컬럼 삭제
df_busan.drop(
    '전출지별',
    axis='columns',
    inplace=True
)

df_busan
```

```python
# 전입지별 -> 전입지
df_busan.rename(
    {'전입지별' : '전입지'},
    axis='columns',
    inplace=True
)

df_busan
```

```python
# 전입지 컬럼을 index로 변경
df_busan.set_index(
    '전입지',
    inplace=True
)
df_busan
```

```python
df_busan.loc['서울특별시', :]
```

```python
# 그래프 사이즈 조정(가로:14인치, 세로:5인치)
plt.figure(
    figsize=(14, 5)
)

# 선그래프
plt.plot(
    df_busan.iloc[0, :]
)

# 차트 제목
plt.title("부산 -> 서울 인구이동")

# 축이름
plt.xlabel('기간')
plt.ylabel('이동 인구수')

# x축 눈금 라벨 회전하기
plt.xticks(
    rotation = 45
)

# y축 눈금 라벨 회전하기
plt.yticks(
    rotation = 45
)

# 범례
plt.legend(
    labels = ['부산 -> 서울'],
    loc = 'best'
    # loc = 'upper left' # upper, lower, left, right
)

plt.show()
```

----
#### 그래프에 스타일 적용하기

```python
# Style 서식 지정
plt.style.use('ggplot')

# 그래프 사이즈 조정(가로:14인치, 세로:5인치)
plt.figure(
    figsize=(14, 5)
)

# 선그래프
plt.plot(
    sr_one,
    marker='o', # o, #, ^, *
    markersize = 12, # marker 크기
    zorder = 1  # Layer의 순서 
)

# 차트 제목
plt.title(
    "서울 -> 경기 인구이동\n", # \n:한줄 띄기
    size=30 # font 크기 
) 

# 축이름
plt.xlabel(
    '\n기간',
    size = 20 # font 크기
)
plt.ylabel(
    '이동 인구수\n',
    size = 20 # font 크기
)

# x축 눈금 라벨 회전하기
plt.xticks(
    rotation = 45
)

# y축 눈금 라벨 회전하기
plt.yticks(
    rotation = 45
)

# 범례
plt.legend(
    labels = ['서울 -> 경기'],
    loc='best',
    fontsize=10
)

# y축 범위 설정
plt.ylim(0, 700000)

# 1995년 데이터 표시하기
plt.annotate(
    f"{sr_one['1995']:,}명", # data
    xy=(1995-1970, sr_one['1995']+40000),
    va='baseline',
    ha='center'
)

# 2017년 데이터 표시하기
plt.annotate(
    f"{sr_one['2017']:,}명", # data
    xy=(2017-1970, sr_one['2017']-40000),
    va='baseline',
    ha='center'
)

# 화살표 annotate
plt.annotate(
    f"{sr_one['1970']:,}명", # data
    xy=(0, sr_one['1970']), # 화살표의 머리부분
    xytext=(1, sr_one['1970']+100000), # 화살표의 몸통부분
    xycoords='data',
    arrowprops=dict(arrowstyle='->', color='blue', lw=3),
    va='baseline',
    ha='center'
)

# 특정 위치의 marker만 틀리게 하기 
plt.scatter(
    1995-1970, # x축
    sr_one['1995'], # y축
    color = 'blue',
    s=500, # size
    marker='o'
)

plt.show()
```

----
#### 화면을 분할하여 그래프 여러개 그리기

```python
# Style
plt.style.use('ggplot')

# Figure
fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(2,1,1) # 2행 1열의 첫번째 
ax2 = fig.add_subplot(2,1,2) # 2행 1열의 두번째 

# ax1
ax1.plot(
    sr_one,
    'o',
    markersize=10
)

# ax2
ax2.plot(
    sr_one,
    marker='o',
    markersize=10,
    markerfacecolor='green', # marker color
    color="#0A743ABC",
    linewidth = 3, # 선 두께 
    label = '서울 -> 경기' # for legend
)

# 범례표시
ax2.legend(loc='best')

# y축 범위 지정
ax2.set_ylim(
    0,
    sr_one.values.max() + 100000
)

# 축눈금 라벨 지정 및 회전
ax2.set_xticks(range(len(sr_one.index))) # x축 갯수 정하기
ax2.set_xticklabels(sr_one.index, rotation=75)

plt.show()
```

----
#### subplot으로 차트 한개 그리기

```python
# Style
plt.style.use('ggplot')

# Figure
fig = plt.figure(figsize=(20, 5))
ax = fig.add_subplot(1,1,1) # 1행1열 첫번째

ax.plot(
    sr_one,
    marker = 'o',
    markersize = 10,
    markerfacecolor='orange',
    color='olive',
    linewidth=2,
    label = '서울 -> 경기'
)

# 범례 표시
ax.legend(loc='best')

# y축 범위 지정
ax.set_ylim(
    0, # 최소값
    sr_one.values.max() + 100000 # 최대값 + 여유분
)

# x축 눈금 라벨 지정 및 회전
ax.set_xticks(range(len(sr_one.index)))
ax.set_xticklabels(sr_one.index, rotation=75)

# 제목
ax.set_title('서울 -> 경기 인구 이동', size=20)

# 축이름
ax.set_xlabel('기간', size=12)
ax.set_ylabel('이동 인구수', size=12)

# 축눈금 라벨 크기
ax.tick_params(axis='x', labelsize=5)
ax.tick_params(axis='y', labelsize=15)

plt.show()
```

----
#### 동일한 그림에 여러개의 그래프를 추가
: 서울에서 충청남도, 경상북도, 강원도로 이동한 인구 데이터의 시각화 

```python
# 서울 -> 충청남도, 경상북도, 강원도 이동 데이터
df_3 = df_seoul.loc[['충청남도', '경상북도', '강원도'],:]
df_3
```

```python
# Style
plt.style.use('ggplot')

# Figure
fig = plt.figure(figsize=(20, 5))
ax = fig.add_subplot(1,1,1) # 1행1열 첫번째

# 충청남도
ax.plot(
    df_3.loc['충청남도',:],
    marker = 'o',
    markersize = 10,
    markerfacecolor='green',
    color='olive',
    linewidth=2,
    label = '서울 -> 충남'
)

# 경상북도 : blue, skyblue
ax.plot(
    df_3.loc['경상북도',:],
    marker='o',
    markersize=10,
    markerfacecolor='blue', # marker color 
    color="skyblue", # line color
    linewidth=2, # line width
    label = '서울->경북'
)

# 강원도 : red, magenta
ax.plot(
    df_3.loc['강원도',:],
    marker='o',
    markersize=10,
    markerfacecolor='red', # marker color 
    color="magenta", # line color
    linewidth=2, # line width
    label = '서울->강원'
)

# 범례 표시
ax.legend(loc='best')

# y축 범위 지정
ax.set_ylim(
    0, # 최소값
    df_3.loc['충청남도',:].max() + 10000 # 최대값 + 여유분
)

# x축 눈금 라벨 지정 및 회전
ax.set_xticks(range(len(df_3.columns)))
ax.set_xticklabels(df_3.columns, rotation=75)

# 제목
ax.set_title('서울 -> 충남, 경북, 강원 인구이동', size=20)

# 축이름
ax.set_xlabel('기간', size=12)
ax.set_ylabel('이동 인구수', size=12)

# 축눈금 라벨 크기
ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)

plt.show()
```
