#### DataFrame
- Pandas에서 사용하는 기본 데이터
- 데이터프레임을 정의할때는 2차원 리스트를 전달 받는다.    
---

#### Pandas를 이용하여 DataFrame 만들기

```python
import pandas as pd
```

```python
# 행방향으로 DataFrame 만들기 : List

# List로 데이터 정의
list01 = [
	[10, 20, 30],
	[40, 50, 60],
	[70, 80, 90]
]
list01

# DataFrame 생성
df = pd.DataFrame(list01)
df
```

```python
# List Type으로 Dataset을 정의하여 DataFrame 생성

df = \
	pd.DataFrame(
		[
			[10,20,30],
			[40,50,60],
			[70,80,90]
		]
	)
df
```

```python
# Column이름이 Pandas의 기본 이름이므로 Column이름 변경하기
df.columns = ['A', 'B', 'C']
df
```

> DataFrame은 index(rows)와 columns로 구성되어 있음.

#### DataFrame에서 값 추출하기

```python
# 열 기준으로 추출하기 :A번 열의 데이터 추출하기
df.loc[:,'A']
```

```python
# 열 기준으로 추출 : A,B번 열 데이터 추출하기

df.loc[:,['A','B']] # column 이름으로 추출
df.loc[:,'A':'B']   # column의 범위로 추출 
df.loc[:,:'B']
```

```python
# 범위를 정해 행 기준으로 추출하기

df.loc[0,:]
```

```python
# 범위를 정해 행 기준으로 추출하기

df.loc[:1,:]  # Slicing
df.loc[[0,1],:]
```
